```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn_pandas import DataFrameMapper
import sklearn


df1 = pd.read_csv('/content/drive/MyDrive/Project/Dataset/edata.csv')
df = pd.read_csv('/content/drive/MyDrive/Project/Dataset/IEB_export.csv')
# Check if the columns exist before deleting them
columns_to_delete = ['Year', 'Month', 'Day', 'Region']
for column in columns_to_delete:
    if column in df.columns:
        del df[column]
    else:
        print(f"Warning: Column '{column}' not found in DataFrame.")
df = df.iloc[::-1]
# df = pd.get_dummies(df)


train_test_ratio = .7
train_size = int(df.shape[0] * train_test_ratio)
train_data = df.iloc[:train_size].copy() # Create a copy to avoid modifying the original DataFrame
test_data = df.iloc[train_size:].copy()  # Create a copy to avoid modifying the original DataFrame


# Apply DataFrameMapper only once to the original DataFrames
mapper = DataFrameMapper([
    # (['Region'], None),
    (['Mag'], sklearn.preprocessing.StandardScaler()),
    (['Lat'], sklearn.preprocessing.StandardScaler()),
    (['Lon'], sklearn.preprocessing.StandardScaler()),
    (['Depth'], sklearn.preprocessing.StandardScaler()),  # Use the correct column name 'Depth km'
    (['Timestamp'], sklearn.preprocessing.StandardScaler())
], default=None)

train_data_transformed = mapper.fit_transform(train_data)
test_data_transformed = mapper.transform(test_data)

# Now, train_data_transformed and test_data_transformed are NumPy arrays
# You can proceed with your model training using these transformed arrays
```

```python
mapper = DataFrameMapper([
       # (['Region'], None),
        (['Mag'],sklearn.preprocessing.StandardScaler()),
        (['Lat'],sklearn.preprocessing.StandardScaler()),
        (['Lon'],sklearn.preprocessing.StandardScaler()),
        (['Depth'],sklearn.preprocessing.StandardScaler()), # Replace 'Depth km' with 'Depth_km'
        (['Timestamp'], sklearn.preprocessing.StandardScaler())
    ],default = None)
train_data = mapper.fit_transform(train_data)
#train_data[['Mag','Lat','Lon','Timestamp','Depth km']] = train_data[['Mag','Lat','Lon','Timestamp','Depth km']].astype(float)
test_data = mapper.transform(test_data)
#test_data[['Mag','Lat','Lon','Timestamp','Depth km']] = test_data[['Mag','Lat','Lon','Timestamp','Depth km']].astype(float)
```

```python
past = 14  # number of past earthquakes to consider
def getXandY(data):
    X, Y = [], []
    for i in range(data.shape[0]-past):
        X += [data[i:i+past].flatten()]
        Y += [data[i+past]]
    return X, Y
train_X, train_Y = getXandY(train_data)
test_X, test_Y = getXandY(test_data)
train_X, train_Y, test_X, test_Y = np.array(train_X), np.array(train_Y), np.array(test_X), np.array(test_Y)
print(np.shape(train_X), np.shape(train_Y))
print(np.shape(test_X), np.shape(test_Y))
```
